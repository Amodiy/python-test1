```python
class phonesimulation:
    def __init__(self,brand,model,colour,year):
        self.brand =brand
        self.model=model
        self.colour=colour
        self.year=year
    def audio(self):
        option = input('Enter audio option')
        if option == 'play':
            print('Audio playing')
        elif option == 'stop':
            print('Audio stopped playing')
phone1= phonesimulation('Techno', 'Canon', 'Black', 2023)
phone1.audio()
```

```python
class anothersimulation(phonesimulation):
    def camera(self):
        option = input('Enter Camera option')
        if option == 'selfie':
            print('Front Rear Camera Activated')
        elif option == 'normal':
            print('Back rear camera activated')
phone2 = anothersimulation('Techno', 'Canon', 'Black', 2024)       
phone2.camera()
```
